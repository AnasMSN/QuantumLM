```python
import os
import shutil
import re

def group_files_by_qubit(source_folder):
    if not os.path.exists(source_folder):
        print(f"Error: The folder '{source_folder}' does not exist.")
        return
    
    # Regular expression to match filenames with 'N<number>' pattern
    pattern = re.compile(r'N(\d+)')
    
    for filename in os.listdir(source_folder):
        match = pattern.search(filename)
        if match:
            qubit_number = match.group(1)  # Extract the number
            target_folder = os.path.join(source_folder, f"{qubit_number}-qubit")
            
            # Create directory if it doesn't exist
            if not os.path.exists(target_folder):
                os.makedirs(target_folder)
            
            # Move file to the target folder
            source_path = os.path.join(source_folder, filename)
            target_path = os.path.join(target_folder, filename)
            shutil.move(source_path, target_path)
            print(f"Moved {filename} -> {target_folder}")

```

```python
source_directory = "generated_images_qutip_grayscale/"
group_files_by_qubit(source_directory)
```
